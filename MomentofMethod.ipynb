```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1.0  # Length of the wire
N = 100  # Number of segments
dx = L / N  # Length of each segment

# Create an array for the center of each segment
x = np.linspace(dx / 2, L - dx / 2, N)

# Initialize the impedance matrix Z and the voltage vector V
Z = np.zeros((N, N))
V = np.ones(N)  # Assume a unit voltage for simplicity

# Fill the impedance matrix Z
for i in range(N):
    for j in range(N):
        if i == j:
            Z[i, j] = np.pi * dx / 2
        else:
            Z[i, j] = dx / np.abs(x[i] - x[j])

# Solve the system of linear equations Z * Q = V to find the charge distribution Q
Q = np.linalg.solve(Z, V)

# Plot the charge distribution
plt.plot(x, Q, label='Charge Distribution')
plt.xlabel('Position along the wire')
plt.ylabel('Charge Density')
plt.title('Charge Distribution on a Thin Wire using the Method of Moments')
plt.legend()
plt.grid(True)
plt.show()
```
